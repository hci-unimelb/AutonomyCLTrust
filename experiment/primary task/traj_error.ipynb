```python
# imports
from os import getcwd
from pandas import read_csv
import matplotlib.pyplot as plt

############################################################# 
### Computes the (mean, standard deviation) of trajectory 
### tracking error under (low, high) autonomy levels
#############################################################
```

```python
file_path = getcwd() + "\data\\traj_preprocessed.csv"
raw_data = read_csv(file_path)
low_auto_df = raw_data[raw_data['auto_grouped']=='low']
low_auto_df = low_auto_df[low_auto_df['autonomy']!=0.4]
high_auto_df = raw_data[raw_data['auto_grouped']=='high']
```

```python
part_id_list = low_auto_df['pid'].tolist()

low_auto_err_list = low_auto_df['traj_err'].tolist()
high_auto_err_list = high_auto_df['traj_err'].tolist()

plt.plot(part_id_list, low_auto_err_list, color='r', label='error (low)')
plt.plot(part_id_list, high_auto_err_list, color='b', label='error (high)')
plt.xlabel("Trajectory ID")
plt.ylabel("RSME (cm)")
plt.title("Error under Low/High Autonomy Across Participants")

plt.legend()
plt.show()
```

```python
# low autonomy
low_mean = sum(low_auto_err_list) / len(low_auto_err_list)
low_var = sum((i - low_mean) ** 2 for i in low_auto_err_list) / len(low_auto_err_list) 

# high autonomy
high_mean = sum(high_auto_err_list) / len(high_auto_err_list)
high_var = sum((i - high_mean) ** 2 for i in high_auto_err_list) / len(high_auto_err_list) 

print("The LOW mean RSME is %.3f cm, with variance = %.3f cm" % (low_mean, low_var))
print("The HIGH mean RSME is %.3f cm, with variance = %.3f cm" % (high_mean, high_var))
```
